```python
import numpy as np
import pandas as pd
from pydataset import data
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from matplotlib import pyplot as plt
```

```python
%matplotlib notebook
pima=data('Pima.tr')
```

```python
pima.plot(kind='scatter',x='skin',y='bmi')
plt.show()
```

```python
x_train,x_test,y_train,y_test=train_test_split(pima.skin,pima.bmi)
```

```python
plt.scatter(x_train,y_train,label='Training data',color='red',alpha=.9)
plt.scatter(x_test,y_test,label='Testing data',color='green',alpha=.9)
plt.legend()
plt.title('training and testing data')
plt.show()
```

```python
LR=LinearRegression()
LR.fit(x_test.values.reshape(-1,1),y_test.values)
```

```python
prediction=LR.predict(x_test.values.reshape(-1,1))
plt.plot(x_test.values.reshape(-1,1),prediction,label='Linear Regression',color='blue',alpha=.9)
plt.scatter(x_test,y_test,label='Actual test data',color='green',alpha=.9)
plt.legend()
plt.title('Final Model')
plt.show()
```

```python
LR.score(x_test.values.reshape(-1,1),y_test.values)
```
